outlier

```python
from scipy.stats import zscore

# Assuming df_clean is your DataFrame after cleaning
numeric_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    # Calculate z-scores
    df_clean[col + '_zscore'] = zscore(df_clean[col])

    # Analyze z-scores (e.g., identify outliers)
    outliers = df_clean[(df_clean[col + '_zscore'] > 3) | (df_clean[col + '_zscore'] < -3)]
    print(f"Outliers in {col}:")
    print(outliers[[col, col + '_zscore']])

    #Visualization
    plt.figure(figsize=(10,6))
    sns.histplot(df_clean[col + '_zscore'], kde=True)
    plt.title(f'Z-score Distribution of {col}')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.show()
```
